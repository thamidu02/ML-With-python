```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine
wine=load_wine()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(wine.data,wine.target,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

neighbor_settings=range(1,11)
training_accuracy=[]
testing_accuracy=[]

for n_neighbors in neighbor_settings:
    wine_model=KNeighborsClassifier(n_neighbors=n_neighbors)
    wine_model.fit(X_train,y_train)
    training_accuracy.append(wine_model.score(X_train,y_train))
    testing_accuracy.append(wine_model.score(X_test,y_test))


plt.plot(neighbor_settings,training_accuracy,label="Training accuracy")
plt.plot(neighbor_settings,testing_accuracy,label="Testing accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
final_wine_model=KNeighborsClassifier(n_neighbors=7)
final_wine_model.fit(X_train,y_train)
```

```python
print("Final model accuracy:",final_wine_model.score(X_test,y_test)*100)
```
