```python
import numpy as np
import matplotlib.pyplot as plt
from websockets.cli import send_outgoing_messages
```

```python
from sklearn.datasets import load_diabetes
diabetes=load_diabetes()
```

```python
diabetes.keys()
```

```python
diabetes.feature_names
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(diabetes.data,diabetes.target,random_state=42)
```

```python
gbrt=GradientBoostingRegressor(n_estimators=100,max_depth=2,learning_rate=1)
gbrt.fit(X_train,y_train)
```

```python
print("Training Accuracy: {:.3f}".format(gbrt.score(X_train,y_train)))
print("Testing Accuracy: {:.2f}".format(gbrt.score(X_test,y_test)))
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=0.01,random_state=0)
ridge.fit(X_train,y_train)
```

```python
print("Training Accuracy: {:.3f}".format(ridge.score(X_train,y_train)))
print("Testing Accuracy: {:.3f}".format(ridge.score(X_test,y_test)))
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_Scaled=scaler.transform(X_test)

mlp=MLPRegressor(solver='lbfgs',random_state=0,hidden_layer_sizes=[10])
mlp.fit(X_train_scaled,y_train)
```

```python
print("Training Accuracy: {:.3f}".format(mlp.score(X_train_scaled,y_train)))
print("Testing accuracy: {:.3f}".format(mlp.score(X_test_Scaled,y_test)))
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=0.0001,max_iter=100000,random_state=0)
lasso.fit(X_train,y_train)
```

```python
print("Training accuracy: {:.3f}".format(lasso.score(X_train,y_train)))
print("Testing Accuracy: {:.3f}".format(lasso.score(X_test,y_test)))
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
print("Training Accuracy: {:.3f}".format(regressor.score(X_train,y_train)))
print("Testing Accuracy: {:.3f}".format(regressor.score(X_test,y_test)))
```
